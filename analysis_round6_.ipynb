```python
# Check the data types of the columns
import numpy as np
x_numer = data.select_dtypes(include=[np.number])

display (x_numer)
```

```python
# Use seaborn library to construct distribution plots for the numerical variables
import seaborn as sns
for cols1 in x_numer.columns:
    sns.displot(x_numer[cols1])
    plt.show()
```

```python
#Use Matplotlib to construct histomgrams
import matplotlib.pyplot as plt
%matplotlib inline

for cols1 in x_numer.columns:
    plt.hist(x_numer[cols1], bins=10)
    plt.xlabel(cols1)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {cols1}')
    plt.show()
```

```python
#Do the distributions for different numerical variables look like a normal distribution

import scipy.stats as stats
import matplotlib.pyplot as plt

# Loop through the numerical columns
for cols1 in x_numer.columns:
    plt.hist(x_numer[cols1], bins=10)
    plt.xlabel(cols1)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {cols1}')
    k2, p = stats.normaltest(x_numer[cols1])
    alpha = 0.05
    if p < alpha:  # null hypothesis: x comes from a normal distribution
        print(f'{cols1} is not normally distributed')
    else:
        print(f'{cols1} is normally distributed')
    plt.show()
```

```python
#For the numerical variables, check the multicollinearity between the features

corr_matrix =x_numer.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
#Drop one of the two features that show a high correlation between them (greater than 0.9). 
# Identify the columns that show a high correlation
high_corr = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr = high_corr.stack().to_frame().reset_index()
high_corr.columns = ['feature1', 'feature2', 'correlation']
high_corr = high_corr.query("correlation > 0.9")

# If any columns show a high correlation, drop one of the columns
if high_corr.shape[0] > 0:
    to_drop = high_corr.feature2.unique()
    x_numer = x_numer.drop(columns=to_drop)
    print(f'The following columns are dropped due to high correlation: {to_drop}')

# Create a heatmap of the updated correlation matrix



sns.heatmap(x_numer.corr(), annot=True)
plt.show()
```
